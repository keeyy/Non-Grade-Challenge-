```python
%matplotlib inline
import random
import matplotlib.pyplot as plt
```

```python
heads_tails = [0,0]
trials = 50000
trial = 0
while trial < trials:
    trial = trial + 1
    toss = random.randint(0, 1)
    heads_tails[toss] = heads_tails[toss] + 1
print(heads_tails)
```

```python
plt.figure(figsize=(5, 5))
plt.pie(heads_tails, autopct = '%.2f%%', labels = ['heads', 'tails'])
plt.legend()
plt.show()
```
